## 1. Scatter plot

Dibuja un scatter plot con 100 puntos distribuidos al azar, tamaños también distribuidos al azar y colores elegidos entre 5 categorias

```python
%matplotlib inline
import random
import numpy
import matplotlib
import matplotlib.pyplot as plt

n_puntos = 100
valores_x = 10 * numpy.random.randn(n_puntos)
valores_y = 10 * numpy.random.randn(n_puntos)

tamanyos = 10 * numpy.random.randn(n_puntos)
tamanyos += abs(tamanyos.min())

posibles_categorias = ['A', 'B', 'C', 'D', 'E']
categorias = numpy.array([random.choice(posibles_categorias) for idx in range(n_puntos)])

for categoria in posibles_categorias:
    mascara_categoria = categorias == categoria
    valores_x_categoria = valores_x[mascara_categoria]
    valores_y_categoria = valores_y[mascara_categoria]
    tamanyos_categoria = tamanyos[mascara_categoria]
    plt.scatter(valores_x_categoria, valores_y_categoria, s=tamanyos_categoria, label=categoria)

plt.legend()
```

## 2. Dos ejes.

Crea una gráfica con dos curvas en la que cada una de las curvas esté asociada a un eje y diferente.

```python
import numpy
import matplotlib.pyplot as plt

# Creamos una exponencial y un seno
t = numpy.linspace(0.01, 10.0, 100)
data1 = numpy.exp(t)
data2 = numpy.sin(2 * numpy.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # el segundo axes comparte el eje y con el primero

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

# 3. Ejes logarítmicos.

Crea un gráfico con un eje y logarítmico para representar una exponencial.

```python
import numpy
import matplotlib.pyplot as plt

# Creamos la exponencial
valores_x = numpy.linspace(0.01, 10.0, 100)
valores_y = numpy.exp(valores_x)

plt.semilogy(valores_x, valores_y)
```

## 4. Boxplot y violín.

Crea un boxplot y un violin plot en el que dibujes 4 distribuciones normales

```python
import random
import numpy
import matplotlib.pyplot as plt

n_distribs = 4
n_muestras = 1000

datos = []
for idx in range(n_distribs):
    media = random.random()
    desviacion = random.random() * 0.2
    muestras = numpy.random.normal(media, desviacion, n_muestras)
    datos.append(muestras)

plt.subplot(2, 1, 1)
plt.boxplot(datos)

plt.subplot(2, 1, 2)
plt.violinplot(datos)
None
```

## 5. Barras apiladas

Dibuja un diagrama de barras apiladas. Cada una de las 5 barras apiladas debe representar las medias para un grupo de personas distinto. La barra inferior representará a los hombres y la superior a las mujeres. La altura de la barra representará la media. Añade también la desviación típica como barra de error.

```python
import random
import numpy
import matplotlib.pyplot as plt

grupos = ['A', 'B', 'C', 'D', 'E']

medias_hombres = []
desviaciones_hombres = []
medias_mujeres = []
desviaciones_mujeres = []
for grupo in grupos:
    medias_hombres.append(random.gauss(1, 0.2))
    desviaciones_hombres.append(random.random() * 0.25)
    medias_mujeres.append(random.gauss(1, 0.2))
    desviaciones_mujeres.append(random.random() * 0.25)

posiciones_x = numpy.arange(len(grupos))

altura = medias_hombres
plt.bar(posiciones_x, altura, label='Hombres', yerr=desviaciones_hombres)

altura = medias_mujeres
bottom = medias_hombres
plt.bar(posiciones_x, altura, bottom=bottom, label='Mujeres', yerr=desviaciones_mujeres)
plt.legend()
```
